```python
cls = ['A', 'B', 'Y']
clss = ['A', 'B', 'Y']
clsss = ['A', 'B', 'Y']
clssss= ['A', 'B', 'Y']
test = '/content/test.data'
train1 = '/content/train1.data'
train2 = '/content/train2.data'
train3 = '/content/train3.data'
```

```python
test = pd.read_csv(test, names=cls)
train1 = pd.read_csv(train1, names=clss)
train2 = pd.read_csv(train2, names=clsss)
train3 = pd.read_csv(train3, names=clssss)
```

```python
data = [train1, train2, train3, test]
data = pd.concat(data)
```

```python
data
```

```python
x=data.iloc[:,0:2].values
```

```python
pd.DataFrame(x).head()
```

```python
y = data.iloc[:,2].values
```

```python
pd.DataFrame(y).head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
from sklearn.svm import SVC;
clf = SVC(kernel='rbf', random_state=0)
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
print("Accuracy :",accuracy_score(y_test,y_pred),"\n")
print("Confusion matrix")
confusion_matrix(y_test,y_pred)
```

```python
plt.scatter(x_test[y_pred==0,0],x_test[y_pred==0,1],color='red');
plt.scatter(x_test[y_pred==1,0],x_test[y_pred==1,1],color='blue');
plt.scatter(x_test[y_pred!=y_test,0],x_test[y_pred!=y_test,1],s=8,color='white');
plt.show()
```
